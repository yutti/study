```python
import pandas as pd
import investpy
import datetime
import os

now = datetime.datetime.now()
today = now.strftime("%d/%m/%Y")

#code_取得
code = '6724' #セイコーエプソン

#フォルダ作成
os.makedirs('log/6724_seiko_epson', exist_ok=True)

'''
#会社業績(会社業績)
term_price    = 'False'
year_price    = 'False'
quarter_price = 'False'
cash_flow     = 'False'
brance_sheet  = 'False'

#世界経済(World economy)
termexchange        = 'False'
economic_indicators = 'False'
'''
#会社業績(会社業績)
term_price    = 'Ture'
year_price    = 'Ture'
quarter_price = 'Ture'
cash_flow     = 'Ture'
brance_sheet  = 'Ture'

performance = ['term_price', 'year_price', 'quarter_price', 'cash_flow', 'brance_sheet']

if term_price:
    stock_data = investpy.get_stock_historical_data(stock=code, country='japan', from_date='01/01/2010', to_date=today)
    #print(stock_data.tail(5))
    #csvの出力
    stock_data.to_csv('log/' + code + '_' + company + '/' + performance[0] +'.csv')

#通年業績
if year_price:
    annual = investpy.get_stock_financial_summary(stock=code, country='japan', summary_type='income_statement',
                                                  period='annual')
    #print(annual.tail(5))
    #csvの出力
    annual.to_csv('log/' + code + '_' + company + '/' + performance[1] +'.csv')
    
#四半期業績
if quarter_price:
    quarterly = investpy.get_stock_financial_summary(stock=code, country='japan', summary_type='income_statement',
                                                     period='quarterly')
    #print(quarterly.tail(5))
    #csvの出力
    quarterly.to_csv('log/' + code + '_' + company + '/' + performance[2] +'.csv')
    
#キャッシュフロー
if cash_flow:
    cash_flow = investpy.get_stock_financial_summary(stock=code, country='japan', summary_type='cash_flow_statement',
                                                     period='annual')
    #print(cash_flow.tail(5))
    #csvの出力
    cash_flow.to_csv('log/' + code + '_' + company + '/' + performance[3] +'.csv')
    
#バランスシート
if brance_sheet:
    brance_sheet = investpy.get_stock_financial_summary(stock=code, country='japan', summary_type='balance_sheet',
                                                         period='annual')
    #print(balance_sheet.tail(5))
    #csvの出力
    brance_sheet.to_csv('log/' + code + '_' + company + '/' + performance[4] +'.csv')
```
